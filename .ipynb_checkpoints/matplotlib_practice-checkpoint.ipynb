```python
%matplotlib inline
```

```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis,y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
plt.plot(x_axis, y_axis, color = "blue", marker = "*", label = "Boston")
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.ylim(0,50)
plt.title("PyBer Fare By Month")
plt.legend()
plt.grid()
```

```python
ax = plt.axes()
ax.plot(x_axis, y_axis, color = "green", marker = "d", label = "Boston")
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
ax.legend()
ax.grid()
ax.set_ylim(0,50)
ax.set_title("PyBer Fare By Month")
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.bar(x_axis,y_axis, color = "green", label = "Boston")
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.title("PyBer Fare by Month")
plt.legend()
```

```python
plt.barh(x_axis,y_axis, color = "magenta", label = "Boston")
plt.ylabel("Date")
plt.xlabel("Fare($)")
plt.gca().invert_yaxis()
plt.legend()
plt.title("PyBer Fare By Month")
```

```python
ax = plt.axes()
ax.barh(x_axis,y_axis, color = "cyan", label = "Chicago")
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
ax.set_title("PyBer Fare By Month")
ax.legend()
```

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
plt.scatter(y_axis,x_axis, color = "red", label = "Chicago", s = y_axis_larger)
plt.legend()
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.xlim(0,50)
plt.gca().invert_yaxis()
```

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*5)
ax = plt.axes()
ax.scatter(y_axis,x_axis, color = "skyblue", edgecolors = "black", alpha = 0.8, label = "Chicago", s = y_axis_larger)
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
ax.set_title("PyBer Fare By Month")
ax.legend()
plt.gca().invert_yaxis()
```
